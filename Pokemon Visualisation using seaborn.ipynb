```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
pokemon=pd.read_csv('Pokemon.csv')
```

```python
pokemon.head()
```

```python
pokemon = pokemon.drop(['Generation', 'Legendary'],1)

```

```python
plt.figure(1,figsize=(15,6))
sns.jointplot('HP','Attack',data=pokemon)
plt.show()
```

```python
sns.boxplot(x='HP',data=pokemon)
```

```python
pokemon = pokemon.drop(['Total', '#'],1)
```

```python
pokemon.head()
```

```python
plt.figure(figsize=(15,7))
sns.boxplot(data=pokemon)
plt.legend()
```

```python
pokemon=pd.melt(pokemon,id_vars=['Name','Type 1','Type 2'],var_name='Stat')
```

```python
pokemon.head()
```

```python
plt.figure(figsize=(15,7))
plt.ylim(0, 275)
sns.swarmplot(data=pokemon,x='Stat',y='value',hue='Type 1')
plt.title('Pokemon Stats')
plt.legend(bbox_to_anchor=(1, 1), loc=2, borderaxespad=0.)
```

```python
sns.set_style("whitegrid")
with sns.color_palette([
    "#8ED752", "#F95643", "#53AFFE", "#C3D221", "#BBBDAF",
    "#AD5CA2", "#F8E64E", "#F0CA42", "#F9AEFE", "#A35449",
    "#FB61B4", "#CDBD72", "#7673DA", "#66EBFF", "#8B76FF",
    "#8E6856", "#C3C1D7", "#75A4F9"], n_colors=18, desat=.9):
    plt.figure(figsize=(15,7))
    plt.ylim(0, 275)
    sns.swarmplot(x="Stat", y="value", data=pokemon, hue="Type 1", split=True, size=7)
    plt.legend(bbox_to_anchor=(1, 1), loc=2, borderaxespad=0.);
```
